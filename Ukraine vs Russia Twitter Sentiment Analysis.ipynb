### Tweetlerin Ã§oÄŸu Ä°ngilizce. Bu verileri duygu analizi gÃ¶revi iÃ§in hazÄ±rlayalÄ±m. Burada tweetlerdeki tÃ¼m baÄŸlantÄ±larÄ±, noktalama iÅŸaretlerini, sembolleri ve diÄŸer dil hatalarÄ±nÄ± kaldÄ±ralÄ±m.

```python
nltk.download('stopwords')
stemmer = nltk.SnowballStemmer("english")
stopword=set(stopwords.words('english'))

def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
df["tweet"] = df["tweet"].apply(clean)
```

```python
text = " ".join(i for i in df.tweet)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### Bu veri kÃ¼mesine tweetlerin duygu puanlarÄ±nÄ± hesaplayarak Pozitif, Negatif ve NÃ¶tr olarak Ã¼Ã§ sÃ¼tun daha ekleyelim:

```python
nltk.download('vader_lexicon')
sentiments = SentimentIntensityAnalyzer()
df["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in df["tweet"]]
df["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in df["tweet"]]
df["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in df["tweet"]]
df = df[["tweet", "Positive", "Negative", "Neutral"]]
print(df.head())
```

### Olumlu duygulara sahip insanlarÄ±n en Ã§ok kullandÄ±ÄŸÄ± kelimelere bir gÃ¶z atalÄ±m:

```python
positive =' '.join([i for i in df['tweet'][df['Positive'] > df["Negative"]]])
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="blue").generate(positive)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### Olumsuz duygulara sahip insanlarÄ±n en Ã§ok kullandÄ±ÄŸÄ± kelimelere bir gÃ¶z atalÄ±m:

```python
negative =' '.join([i for i in df['tweet'][df['Negative'] > df["Positive"]]])
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="red").generate(negative)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
